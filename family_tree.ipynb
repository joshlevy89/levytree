```python
class Person:

    def __init__(self):

        self.name = None
        self.identifier = None
        
def married_name_parser(line):
    def is_int(c):
        try:
            int(c)
            return True
        except:
            return False
    abc = line.split('m: ')[1]
    if is_int(abc[0]):
        xyz = ' '.join(abc.split()[1:]).split(' b:')[0]
    else:
        xyz = abc.split(' b:')[0]
    
    return xyz.replace('(', '\'').replace(')', '\'')

tree = {}
persons = {}
spouses = {}
# TODO: B2 side 
with open('/Users/levyjos/Desktop/family_tree/levy_family_tree.txt', encoding="ISO-8859-1") as f:
    for line in f:
        line = line.strip()
        if line.startswith('B1'):
            identifier = line.split('. ')[0]
            name = line.split('. ')[1].split(' b:')[0]
            name = name.replace('(', '\'').replace(')', '\'')
            person = Person()
            person.name = name
            person.identifier = identifier
            persons[identifier] = person
            if identifier not in tree:
                tree[identifier] = []
            if identifier[:-1] in tree:
                tree[identifier[:-1]].append(person)       
        elif line.startswith('m'):
            married_name = married_name_parser(line)
            spouse = Person()
            spouse.name = married_name
            spouse.identifier = identifier
            spouses[identifier] = spouse

with open('/Users/levyjos/Desktop/family_tree/formatted_levy_family_tree.txt', 'w') as f:
    for parent_id in tree:
        if parent_id not in spouses:
            continue
        
        f.write('\n{} (id={})'.format(persons[parent_id].name, parent_id))
        f.write('\n{} (id=m{})'.format(spouses[parent_id].name, parent_id))            
        if tree[parent_id] == []:
            f.write('\n\t{} (id=nokids{})'.format('no kids', parent_id))
        else:
            for children in tree[parent_id]:
                f.write('\n\t{} (id={})'.format(children.name, children.identifier))
        f.write('\n')
```
